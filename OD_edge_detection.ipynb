```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edge = cv2.Canny(img,127,255)
```

```python
plt.imshow(edge)
```

```python
med_val = np.median(img)
lower = int(max(0,0.7*med_val))
upper = int(min(255,1.3*med_val))
```

```python
edge = cv2.Canny(img,lower,upper)
```

```python
plt.imshow(edge)
```

```python
blurred_img = cv2.blur(img,(5,5))
edge = cv2.Canny(blurred_img,lower,upper+50)
```

```python
plt.imshow(edge)
```
