```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_pic(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('DATA/sudoku.jpg', 0)
```

```python
plt.imshow(img,cmap='gray')
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
display_pic(sobelx)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_pic(laplacian)
```

```python
#combining the bot gradients
blended = cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobelx, beta=0.5, gamma=0)
```

```python
display_pic(blended)
```

```python
ret, th1 = cv2.threshold(img,120,255,cv2.THRESH_BINARY)
```

```python
display_pic(th1)
```

```python
kernel = np.ones((4,4), np.uint8)
```

```python
gradient = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT, kernel)
```

```python
display_pic(gradient)
```
