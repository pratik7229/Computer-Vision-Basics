```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_horse = cv2.imread('DATA/horse.jpg')
show_horse = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

dark_horse = cv2.imread('DATA/rainbow.jpg')
show_rainbow = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)

dark_horse = cv2.imread('DATA/bricks.jpg')
show_bricks = cv2.cvtColor(dark_horse, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_horse)
```

```python
hist = cv2.calcHist([show_horse], channels=[0], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.imshow(show_horse)
```

```python
plt.plot(hist) # most of the vaules are near to zero
```

```python
img = show_bricks
color = ('b','g','r')

for i, col in enumerate(color):
    histr =cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
    
plt.title('histogram for bricks')
```

```python
plt.imshow(show_bricks)
```

```python
gorilla = cv2.imread('DATA/gorilla.jpg')
show_gorilla = cv2.cvtColor(gorilla,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(show_gorilla)
```

```python
# for using equalize function we need to convert it to hsv scale
hsv = cv2.cvtColor(gorilla,cv2.COLOR_BGR2HSV)
hsv[:,:,2] = cv2.equalizeHist(hsv[:,:,2])
```

```python
eq_color_gorilla= cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(eq_color_gorilla)
```
